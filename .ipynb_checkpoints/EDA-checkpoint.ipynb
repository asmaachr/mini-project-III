## Fixing Dates & Times ##

```python
#converting time string to three separate variables to then concatenate and add to tran_date to get around 0 times
def datetimemaker(df):
    df['hour'] = df['tran_time']//10000
    df['minute'] = df['tran_time']//100-df['hour']*100
    df['second'] = df['tran_time']%100

    df['time'] = df['tran_date'].astype('str')+" "+df['hour'].astype('str')+":"+df['minute'].astype('str')+":"+df['second'].astype('str')
    df['time'] = pd.to_datetime(df['time'], format = '%d.%m.%Y %H:%M:%S')
    df = df.drop(columns=['tran_date','tran_time'])
```

```python
datetimemaker(transactions)
```

```python
transactions = transactions.drop(columns=['tran_date','tran_time'])
```

```python
def start_end_convert(df):
    df['acct_start_date'] = pd.to_datetime(df['acct_start_date'], format = "%d.%m.%Y")
    df['acct_end_date'] = pd.to_datetime(df['acct_end_date'], format = "%d.%m.%Y")
```

```python
#converted account start date and account end date to datetimes across all 4 df
start_end_convert(accounts)
start_end_convert(checking_acct)
start_end_convert(credit_acct)
start_end_convert(savings_acct)
```
